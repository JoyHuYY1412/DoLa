```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# file_path = './output-path-tfmc-dola-all-layer.json'
file_path = ['mmlu_results/mmlu-college_mathematics-dola-all-layer.json',
            'mmlu_results/mmlu-elementary_mathematics-dola-all-layer.json',
            'mmlu_results/mmlu-high_school_geography-dola-all-layer.json',
            'mmlu_results/mmlu-nutrition-dola-all-layer.json',
             'mmlu_results/mmlu-philosophy-dola-all-layer.json'
            ]
layers_info = []
for i in range(len(file_path)):
    # file_path = './output-path-tfmc-dola.json'
    with open(file_path[i], 'r', encoding='utf-8') as file:
        data = json.load(file)
        layers_info.append(data['premature_layer_dist_correct'])
```

```python
# Convert list of dicts to DataFrame
subjects = ["college_mathematics", "elementary_mathematics", "high_school_geography", "nutrition", "philosophy"]
df = pd.DataFrame(layers_info).T
df.columns = subjects

# Normalize each column independently to [0,1]
df_normalized = df.apply(lambda x: (x - x.min()) / (x.max() - x.min()), axis=0)

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df_normalized, cmap="coolwarm", linewidths=0.5, linecolor='gray', cbar=True, annot=False)
plt.xlabel("Subjects")
plt.ylabel("Layer Index (Normalized)")
plt.title("Heatmap of Values Across Subjects")
plt.show()
```
